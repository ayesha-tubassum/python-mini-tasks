Question 2:

Ek simple library system banayein jisme:
Books ki list create karein.
User ko option mile:
Add book
Remove book
Search book
Display all books
Agar book exist nahi karti to proper message show karein.
Ek function likhein jo total books count kare.

📌 Conditions:

Lists aur functions ka proper use karein.
Menu-driven program ho.

```python
library = ["Python Crash Course", "Harry Potter", "The Alchemist",
           "Atomic Habits", "Clean Code", "Data Science Handbook",
           "Think Like a Programmer", "Deep Work", "The Great Gatsby",
           "Rich Dad Poor Dad", "Sapiens", "1984",
           "To Kill a Mockingbird", "The Pragmatic Programmer", "Algorithms Unlocked"]
my_list = []
def search_book(name):
    for book in library:
        if name.lower() in book.lower():
            print("Found:", book)
            return
    print("Book not found in library.")
def add_book(name):
    if name in library:
        my_list.append(name)
        print(name, "added to your list!")
    else:
        print("This book is not in the library.")
def remove_book(name):
    if name in my_list:
        my_list.remove(name)
        print(name, "removed from your list!")
    else:
        print("Book not found in your list.")
def show_my_list():
    if my_list:
        print("Your List:", my_list)
    else:
        print("Your list is empty!")
def show_library():
    print("Library Books:", library)
while True:
    print("\n1-Search  2-Add  3-Remove  4-My List  5-All Books  6-Exit")
    choice = input("Enter choice (1-6): ")
    if choice == "1":
        search_book(input("Book name: "))
    elif choice == "2":
        add_book(input("Book name to add: "))
    elif choice == "3":
        remove_book(input("Book name to remove: "))
    elif choice == "4":
        show_my_list()
    elif choice == "5":
        show_library()
    elif choice == "6":
        print("Goodbye!")
        break
```

Neeche diye gaye problems recursively solve karein:
Factorial function likhein.

Ek recursive function likhein jo list ka sum calculate kare.
Ek function likhein jo string palindrome hai ya nahi check kare (recursively).
Ek power function likhein jo a^b calculate kare without using **.

📌 Condition:

Har function mein recursion ka proper base case hona chahiye.

```python
#(simple factorial code):
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n-1)

n = int(input("Enter number: "))
print("factorial: ", factorial(n))
```

```python
#(fibanocci sequence):
def fibonacci(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fibonacci(n-1) + fibonacci(n-2)
n = int(input("Enter term number: "))
print("Fibonacci:", fibonacci(n))
```

```python
#(sum of list):
my_list = []
while True:
    number = input("Enter number (or 'done' to stop): ")
    if number == "done":
        break
    my_list.append(int(number))
print("Your list:", my_list)
def list_sum(lst):
    if lst == []:
        return 0
    return lst[0] + list_sum(lst[1:])
print("Sum:", list_sum(my_list))
```

```python
#(palindromic sequence)
def is_palindrome(s):
    if len(s) <= 1:
        return True
    if s[0] != s[-1]:
        return False
    return is_palindrome(s[1:-1])
word = input("Enter word: ")
print("Palindrome:", is_palindrome(word))
```

```python
#(sqaure without using **)
def power(a, b):
    if b == 0:
        return 1
    return a * power(a, b-1)
a = int(input("Enter base: "))
b = int(input("Enter power: "))
print(power(a, b))
```

Question 3:

Ek class BankAccount design karein jisme:
Attributes:
account_holder
__balance (private variable)
Methods:
deposit(amount)
withdraw(amount)
check_balance()
transfer(other_account, amount)
Program mein:
Do accounts create karein.
Money transfer simulate karein.
Withdraw balance se zyada na ho.

📌 Conditions:
Balance private hona chahiye.
Getter aur setter implement karein.
Transaction history maintain karein (list use karein).

```python
class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.__balance = balance
        self.history = []
    def get_balance(self):
        return self.__balance
    def set_balance(self, amount):
        self.__balance = amount
    def deposit(self, amount):
        self.__balance += amount
        self.history.append("Deposit: " + str(amount))
    def withdraw(self, amount):
        if amount > self.__balance:
            print("Not enough balance!")
        else:
            self.__balance -= amount
            self.history.append("Withdraw: " + str(amount))
    def check_balance(self):
        print(self.account_holder, "Balance:", self.__balance)
    def transfer(self, other_account, amount):
        if amount > self.__balance:
            print("Not enough balance!")
        else:
            self.__balance -= amount
            other_account.set_balance(other_account.get_balance() + amount)
            self.history.append("Transfer to " + other_account.account_holder + ": " + str(amount))
account1 = BankAccount("Ayesha", 5000)
account2 = BankAccount("tabassum", 2000)
account1.transfer(account2, 5000)
account1.check_balance()
account2.check_balance()
print("*"*30)
print("Transaction History:")
for transaction in account1.history:
    print(transaction)
```

Question 4:

Ek base class Person banayein jisme:
Attributes:
name
age
Method:
get_details()
Do child classes banayein:
Student (additional attribute: GPA)
Teacher (additional attribute: salary)
Tasks:
get_details() method override karein.
Ek list mein mixed objects store karein.
Loop chala kar polymorphism demonstrate karein.

📌 Condition:

Super() ka use karein.
Method overriding clear hona chahiye.

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def get_details(self):
        print("Name:", self.name, "\nAge:", self.age)
class Student(Person):
    def __init__(self, name, age, gpa):
        super().__init__(name, age)
        self.gpa = gpa
    def get_details(self):
        print("\nName:", self.name, "\nAge:", self.age, "\nGPA:", self.gpa)
class Teacher(Person):
    def __init__(self, name, age, salary):
        super().__init__(name, age)
        self.salary = salary
    def get_details(self):
        print("\nName:", self.name, "\nAge:", self.age, "\nSalary:", self.salary)
people = [
    Student("Ayesha", 20, 3.6),
    Teacher("tabassum", 35, 50000),
    Student("Hamza", 22, 3.8)
]
for person in people:
    person.get_details()
```

Question 5:

Abstract class Shape banayein jisme abstract method ho:
area()
Child classes banayein:
Circle
Rectangle
Triangle
Tasks:
User se input le kar shape select karein.
Required parameters input karein.
Area calculate karein.
Multiple shapes ko list mein store karein aur total area calculate karein.

📌 Condition:
abc module use karein.
Proper abstraction follow karein.

```python
from abc import ABC, abstractmethod
import math
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return math.pi * self.radius * self.radius
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width
class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height
    def area(self):
        return 0.5 * self.base * self.height
shapes = []
while True:
    print("\n1-Circle  2-Rectangle  3-Triangle  4-Total Area  5-Exit")
    choice = input("Choose: ")
    if choice == "1":
        r = float(input("Enter radius: "))
        shapes.append(Circle(r))
        print("Circle area:", Circle(r).area())
    elif choice == "2":
        l = float(input("Enter length: "))
        w = float(input("Enter width: "))
        shapes.append(Rectangle(l, w))
        print("Rectangle area:", Rectangle(l, w).area())
    elif choice == "3":
        b = float(input("Enter base: "))
        h = float(input("Enter height: "))
        shapes.append(Triangle(b, h))
        print("Triangle area:", Triangle(b, h).area())
    elif choice == "4":
        total = 0
        for shape in shapes:
            total += shape.area()
        print("Total area:", round(total, 2))
    elif choice == "5":
        print("Goodbye!")
        break
    else:
        print("Invalid choice!")
```


**PROJECT 1 – Simple Data Analysis Dashboard (Without Libraries)
Problem Statement**

Question 1:

Ek “Mini Data Analysis Dashboard” banayein jo numerical dataset ko analyze kare.
Dataset manually define karein (example: sales data of 12 months).

📝 Project Requirements:
Ek class DataAnalyzer banayein jisme:
data (list of numbers)
Methods implement karein:
calculate_mean()
calculate_median()
calculate_max()
calculate_min()
count_above_average()
find_trend() (increasing, decreasing ya mixed)
Ek menu-driven interface banayein jisme:
User analysis option choose kare.
Har option ka result show ho.

📌 Bonus:
User new number add kar sake.
Data update hone ke baad analytics dobara run kare.

```python
class DataAnalyzer:
    def __init__(self):
        self.data = [1500, 2000, 1800, 2200, 1900, 2100,
                     2300, 2500, 2100, 1950, 2400, 2600]
    def calculate_mean(self):
        return sum(self.data) / len(self.data)
    def calculate_median(self):
        sorted_data = sorted(self.data)
        n = len(sorted_data)
        mid = n // 2
        if n % 2 == 0:
            return (sorted_data[mid - 1] + sorted_data[mid]) / 2
        return sorted_data[mid]
    def calculate_max(self):
        return max(self.data)
    def calculate_min(self):
        return min(self.data)
    def count_above_average(self):
        avg = self.calculate_mean()
        count = 0
        for num in self.data:
            if num > avg:
                count += 1
        return count
    def find_trend(self):
        if self.data[-1] > self.data[0]:
            return "Increasing"
        elif self.data[-1] < self.data[0]:
            return "Decreasing"
        else:
            return "Mixed"
    def add_number(self, num):
        self.data.append(num)
        print("Data added!")
analyzer = DataAnalyzer()
while True:
    print("1. Mean")
    print("2. Median")
    print("3. Maximum")
    print("4. Minimum")
    print("5. Count Above Average")
    print("6. Find Trend")
    print("7. Add New Number")
    print("8. Exit")
    choice = input("Choose (1-8): ")
    if not choice.isdigit() or int(choice) < 1 or int(choice) > 8:
        print("Invalid! Enter 1 to 8 only.")
        continue
    if choice == "1":
        print("Mean:", analyzer.calculate_mean())
    elif choice == "2":
        print("Median:", analyzer.calculate_median())
    elif choice == "3":
        print("Max:", analyzer.calculate_max())
    elif choice == "4":
        print("Min:", analyzer.calculate_min())
    elif choice == "5":
        print("Above Average Count:", analyzer.count_above_average())
    elif choice == "6":
        print("Trend:", analyzer.find_trend())
    elif choice == "7":
        num = float(input("Enter new number: "))
        analyzer.add_number(num)
        print("Updated Data:", analyzer.data)
    elif choice == "8":
        print("Goodbye!")
        break
```
